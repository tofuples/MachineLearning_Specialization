```python
import numpy as np
import matplotlib.pyplot as plt
import matplotx

#plt.style.use(matplotx.styles.aura["dark"])
```

- Learn how to implement the model $f_{w,b} = wx + b $ for linear regression with one variable. 



```python
# x_train is the input variable (size in 1000 square feet)
# y_train is the target (price in 1000s of dollars)

x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# m is the number of training examples

print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0] #lenght of the array
print(f"Number of training examples is: {m}")
```

```python
i = 0

x_i = x_train[i]
y_i = y_train[i]

print(f"x^({i}), y^({i}) = ({x_i}, {y_i}))")
```

Plotting the data

```python
# plot the data points
plt.scatter(x_train, y_train, marker='x')

#title
plt.title("Housing Prices")

# x-axis and y-axis label
plt.ylabel("Prices(1000s of dollars)")
plt.xlabel("Size (1000 sqrt)")

plt.show()
```

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b 
        
    return f_wb
    
```

x $\rightarrow $ data, n-dimensional array of shape (m,), m examples $\\$ 

w, b $\rightarrow $ model parameters $\\$

y $\rightarrow $ target values


```python
w = 200
b = 100
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)


# plotting our model prediction
plt.plot(x_train, tmp_f_wb, label='Our Prediction')

# data points
plt.scatter(x_train, y_train, marker='o', label='Actual Values')

plt.title("Housing Prices")
plt.ylabel("Prices(1000s of dollars)")
plt.xlabel("Size (1000 sqrt)")
plt.legend()

plt.show()
```
